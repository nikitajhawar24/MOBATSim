```python
import pandas as pd
import os
import glob
  
  
# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
  
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
    df.columns = ["a","b","c","d","e","f","g","h","i"]
    df=df.drop(["c","d","e","h"],axis=1)
    df = df.rename(columns={"a": "Time", "b": "Speed", "f":"Rotation", "g":"Position", "i":"Translation"})
    df.to_csv(f,index=False)
    display(df)
    print(f)
```

```python
import numpy as np
data = np.genfromtxt("Road Merge Collision_Noise_faultduration_5_V6.csv", delimiter=",", names=["Time", "Speed","Rotation","Position","Translation"])
data
import matplotlib.pyplot as plt
plt.figure(figsize=(7, 7))
plt.plot(data['Time'], data['Speed'], label='Speed')
plt.plot(data['Time'], data['Rotation'], label='Rotation')
plt.plot(data['Time'], data['Position'], label='Position')
plt.plot(data['Time'], data['Translation'], label='Translation')
plt.xlabel('Time')
plt.ylabel('Data')
plt.title('Vehicle 6')
plt.grid(True)
plt.legend(title='Data',loc='lower right')
plt.show()
```
