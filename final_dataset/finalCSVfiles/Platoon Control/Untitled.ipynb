```python
import pandas as pd
import os
import glob
  
  
# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
  
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
    df.columns = ["a","b","c","d","e","f","g","h","i"]
    df=df.drop(["c","d","e","h"],axis=1)
    df = df.rename(columns={"a": "Time", "b": "Speed", "f":"Rotation", "g":"Position", "i":"Translation"})
    df.to_csv(f,index=False)
    display(df)
    print(f)
```
